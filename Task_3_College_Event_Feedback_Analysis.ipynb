```python
import pandas as pd

# Load dataset
df = pd.read_csv("student_feedback.csv")

# Show first 5 rows
df.head()
# Check dataset structure
df.info()

# Column names
df.columns

# Drop unnecessary columns
df = df.drop(columns=['Unnamed: 0'])

# Verify
df.head()

# Calculate average rating for each feedback question
avg_scores = df.drop(columns=['Student ID']).mean().sort_values(ascending=False)

avg_scores

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.barplot(
    x=avg_scores.values,
    y=avg_scores.index,
    palette="Blues_r"
)

plt.title("Average Student Satisfaction Scores")
plt.xlabel("Average Rating")
plt.ylabel("Feedback Category")
plt.show()

overall_satisfaction = avg_scores.mean()
overall_satisfaction

```
